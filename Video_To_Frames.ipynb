```python
import shutil
import os
from google.colab import files

# Variabile per pulire la cartella relativa alle immagini caricate.
# Se si vogliono cancellare immagini caricate gi√† in precedenza,
# impostare la variabile 'delete' a 'True'.
delete = False

# Se per qualche motivo, questa cella viene avviata all'interno della cartella
# principale del progetto, spostiamoci una cartella prima ("/content").
if(os.getcwd().endswith("Uploads")):
    examplesDirectory = "examples/"
else:
    examplesDirectory = "Uploads/examples/"

if (delete):
  if os.path.exists(examplesDirectory):
    shutil.rmtree(examplesDirectory)
  else:
    print("La cartella dei Video non esiste.\nMolto probabilmente stai runnando la cella per la prima volta.\n")

# Se non esiste la cartella dei video, creiamola
if not os.path.exists(examplesDirectory):
    os.makedirs(examplesDirectory)

# Caricamento dei video
print("Caricamento Video ...")
uploaded = files.upload()

# Variabile per salvataggio dei nomi dei video caricati
filenames = uploaded.keys()

# Per ogni video caricato, spostiamola nella cartella dei video
for f in filenames:
    shutil.move(f, examplesDirectory + f)

print("Caricamento Video completato.")
```

```python
import cv2
from google.colab import drive

video_path = '/content/Uploads/examples/Canvas Glimmer.mp4'
video = cv2.VideoCapture(video_path)
success = True
count = 1
while success:
  success,frame = video.read()
  name = '/content/Results/'+str(count)+'.jpg'
  if success == True:
    cv2.imwrite(name,frame)
    print('Frame {} Extracted Successfully'.format(count))
    count = count+1
  else:
    break

print('Done!')
```
